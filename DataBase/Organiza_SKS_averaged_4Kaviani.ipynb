```python
import pandas as pd
import numpy as np
from scipy.stats import circmean
```

```python
def organizaSTA(station):
    STAlist=[]
    for i,sta in enumerate(station):
        if i ==0: 
            STAlist.append(sta)
            temp_STA=sta
        elif temp_STA!=sta:
            STAlist.append(sta)
            temp_STA=sta
    return STAlist
```

```python
DIR='/Users/tong/Documents/Master/Map_2021/DataBase'
file=f'{DIR}/Kaviani_etal_dead_sea_fault_desert.csv'
df=pd.read_csv(file)
```

```python
station=df['Station']
STAlist=organizaSTA(station)
```

```python
STAlonlist=[];STAlatlist=[];phimeanlist=[];dtmeanlist=[]
for STA in STAlist:
    print(STA)
    stlon=df['Sta_Lon'][df['Station']==STA].values[0]
    stlat=df['Sta_Lat'][df['Station']==STA].values[0]
    phi=(df['Phi(Â°)'][df['Station']==STA]).tolist();
    dt=(df['dt (s)'][df['Station']==STA]).tolist();
    if 'Null' in phi:
        phi=[x for x in phi if x != 'Null']
        dt=[x for x in dt if x != 'Null']
    phi=list(map(float, phi))
    dt=list(map(float, dt))
    phi_mean=np.mean(phi)
    dt_mean=np.mean(dt)
    STAlonlist.append(stlon)
    STAlatlist.append(stlat)
    phimeanlist.append(phi_mean)
    dtmeanlist.append(dt_mean)

newdf=pd.DataFrame({'Station': STAlist,
                   'Sta_Lat':STAlatlist,
                   'Sta_Lon':STAlonlist,
                   'Phi':phimeanlist,
                   'dt':dtmeanlist})
newdf.to_csv(f'{DIR}/Kaviani_etal_dead_sea_fault_desert_averaged.csv',index=False)
```
